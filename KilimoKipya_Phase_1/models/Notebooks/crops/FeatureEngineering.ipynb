```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import joblib as jb
```

```python
data = pd.read_csv("../../Data/crop_recommendation.csv")
data_en = data.drop("label",axis=1)
data_en
```

```python
scaler = StandardScaler()
array_scaled = scaler.fit_transform(data_en)
data_scaled = pd.DataFrame(array_scaled,columns=data_en.columns)
data_scaled
```

```python
data_scaled["label"] = data["label"]
data_scaled.to_csv("output/processed_data/data_scaled.csv",index=False)
data_scaled
```

```python
jb.dump(scaler,"output/model/scaler.joblib")
```
