```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import joblib as jb
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
data = pd.read_csv("../../Data/Fertilizer Prediction.csv")
data_en = data.drop("Fertilizer Name",axis=1)
data_en
```

```python
data_en.columns
```

```python

columns_to_encode = ['Soil Type', 'Crop Type']
columns_to_scale = [col for col in data_en.columns if col not in columns_to_encode]

# Define transformers for encoding and scaling
encoder = OneHotEncoder(handle_unknown='ignore')
scaler = StandardScaler()

# Create a ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('scaler', scaler, columns_to_scale),
        ('onehot', encoder, columns_to_encode)
    ])

# Fit and transform the data
transformed_data = preprocessor.fit_transform(data_en)

# Convert the transformed data to a DataFrame
columns_encoded = preprocessor.named_transformers_['onehot'].get_feature_names_out(columns_to_encode)
columns_scaled = columns_to_scale
columns_transformed = columns_scaled +  list(columns_encoded) 
data_transformed = pd.DataFrame(transformed_data, columns=columns_transformed)

data_transformed
```

```python
data_transformed.columns
```

```python
jb.dump(preprocessor,"output/model/preprocessor.joblib")
```

```python
data_transformed["Fertilizer Name"] = data["Fertilizer Name"]
data_transformed.to_csv("output/processed_data/data_scaled.csv",index=False)
data_transformed
```
